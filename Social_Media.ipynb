# Importing The Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Uploading the dataset

```python
df=pd.read_csv('/content/Time-Wasters on Social Media.csv')
df
```

```python
# prompt: scatter plot of device type and addiction level

plt.figure(figsize=(10, 6))
sns.scatterplot(x='DeviceType', y='Addiction Level', data=df)
plt.title('Scatter Plot of Device Type vs Addiction Level')
plt.xlabel('Device Type')
plt.ylabel('Addiction Level')
plt.xticks(rotation=45)
plt.show()
```

# Usage as per Gender

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Gender'], kde=False)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Male users are more than female users

# Usage as per Demographic

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Demographics'], kde=False)
plt.title('Demographics Distribution')
plt.xlabel('Demographics')
plt.ylabel('Count')
plt.show()
```

Rural users are overwhelmingly more than urban users

# Usage per Profession

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Profession'], kde=False, bins=10)
plt.title('Profession Usage Distribution')
plt.xlabel('Profession')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Students, Waiting staff and Labor workers are the highest users

# Demographic Distribution by Age

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Demographic Distributio (Age)')
plt.xlabel('Age')
plt.ylabel('Frenquency')
plt.show()
```

Demographic Distribution by Age has a uniform distribution

# Watch reason histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Watch Reason'], kde=False, bins=10)
plt.title('Reason for using social media Distribution')
plt.xlabel('Reason for using social media')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Most people use social media majorly out of habit,then boredom and entertainment reasons

#Current Activity During Usage Distribution

```python
plt.figure(figsize=(8, 8))
df['CurrentActivity'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Current Activity Distribution')
plt.ylabel('')  # Hide the y-label
plt.show()
```

majority of the users are online while at home and students in school

# Watch Time Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Watch Time'], kde=False, bins=10)
plt.title('Watch Time Distribution')
plt.xlabel('Watch Time')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

most users are online during their lunch break and after work.
